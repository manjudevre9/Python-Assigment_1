```python
Que.2.Load the "fmrl" dataset using the load_dataset function of seaborn.Plot a line plot using x="timepoint" & 
      y="singal" for different events & regions.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fmr1=sns.load_dataset("fmr1")
sns.lineplot(data=fmr1, x="timepoint", y="single", hue="event", style="regoin")

plt.title("FMR1 Dataset Line Plot")
plt.xlabe("Timepoint")
plt.ylable("Single")

plt.show()
```

```python
Que.3. Load the "titanic" dataset using the load_dataset function of seaborn.Plot two box plots using x='pclass',
       y='age' & y='fare'. 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic=sns.load_dataset("titanic")
fig,(ax1, ax2)=plt.subplots(1,2,figsize=(12,6))
sns.boxplot(x="pclass",y="age",data=titanic,ax=ax1)
ax1.set_time("Age by Class")
ax1.set_xlable("Class")
ax1.set_ylable("Age")

sns.boxplot(x="pclass",y="fare",data=titanic,ax=ax2)
ax2.set_title("Fare by Class")
ax2.set_xlable("Class")
ax2.set_ylable("Fare")

plt.show()
```

```python
Que.4.Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for
      the 'cut' column of the diamonds dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

diamonds=sns.load_dataset("diamonds")

sns.histplot(data=diamonds,x="price",hue="cut",multiple="stack",bins=20)

plt.title("Diamond Price Distribution by Cut")
plt.xlable("Price")
plt.ylable("Count")

plt.show()
```
