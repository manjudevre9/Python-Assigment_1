```python
Que.1.Define overfitting & underfitting machine learning.What are the consequences of each,& how can they be 
      mitigated?
Ans. Overfitting:Overfitting occurs when a machine learning model i too complex & learns the noise in the
     training data,resulting in poor performance on new,unseen data.
    
    Consequences:
        -poor generalization to new data.
        -high variance in prediction.
        -model is sensitive to small changes in input data.
        
    Mitigation strategies:
        -regularization techniques(L1,L2,dropout)
        -early stopping
        -data augmentation
        -reducing model complexity
        -increasing training data size
        
    Underfitting:Underfitting occurs when a machine learning model is too simple & fails to capture the 
    underlying patterns in the training data,resulting in poor performance on both training & new data.
    
    Consequences:
        -poor performanceon training data.
        -high bias in predictions.
        -model is unable to capture important relationship.
        
    Mitigation strategies: 
        -increasing model complexity.
        -adding more features.
        -using more advanced models(e.g.,neural networks)
        -increasing training time.
        -feature engineering.
```

```python
Que.2.How can we reduce overfitting?Exaplain in brief.
Ans.1.Regularization:Add a penalty term to the loss function to discourage large weights.
    2.Dropout:Randomly drop out neurons during training to prevent over-reliance on any single reuron.
    3.Early Stopping:Stop training when performance on the validation set starts to degrade.
    4.Data Augmentation:Increase the size of the training set by generating new data through transformations.
    5.Silplifying the Model:Reduce the complexity of the model by removing unnecessary layers or features.
    6.Ensemble Methods:Combine the predictions of multiple models to reduce variance.
    7.Cross-Validation:Evaluate the model on unseen data to estimate its generalization performance.
    8.L1 & L2 Regularization:Add a penalty term to the loss function to reduce the magnitude of weights.
    9.Batch Normalization:Normalize inputs to each layer to reduce internal covariate shift.
    10.Pruning:Remove unnecessary connections or neurons after training.
```

```python
Que.3.Explain underfitting. List scenario where underfitting can occur in ML.
Ans.Underfitting:Underfitting occurs when a machine learning model is too simple or has too few parameters to 
                 capture the underlying patterns in the training data.As a result the model fails to learn the 
                 relationships between the features & the target variable,leading to poor performance on both 
                 the training & test data.
    
    Scenario where underfitting can occur:
        1.Using a linear model for a non-linear problem.
        2.Using too few features.
        3.Using a model with too few parameters.
        4.Not training the model for enough time.
        5.Using a simple model for a complex & requires a more sophisticated model,using a simple model will 
          lead to underfitting.
```

```python
Que.4.Explain the bias variance tradeoff in machine learning.What is the relationship between bias & variance &
      how do they affect model performance.
Ans.Bias:
    -Bias refers to the difference between the models expected predictions & the true values.
    -High bias means the model is oversimplifying the model is oversimplifying the data missing important 
     patterns.
    -Low bias means the model is capturing the underlying patterns in the data.
    
    Variance:
    -Variance refers to the spread or dispersion of the models predictions.
    -High variance means the model is sensitive to small changes in the data & is overfitting.
    -Low variance means the model is stable & consistent its predictions.
    
    Relationship between Bias & Variance:
        -As bias decreases (model becomes more coplex),variance increses (model becomes more sensitive to data).
        -As variance decreases (model becomes simpler),bias increases (model misses important patterns).
       
    How Bias & Variance affect Model Performance:
        -High Bias & Low Variance:Model is oversimplified,misses important patterns,& performs poorly on both
         training & test data.
        -Low Bias & High Variance:Model is overfitting,performs well on training data but poorly on test data.
        -Optimal Balance:Model has a balance of bias & variance,performs well on both training & test data.
        
```

```python
Que.5.Discuss some common method for defecting overfitting & underfitting in machine learning model. How can 
      you determine whether your model is overfitting or underfitting.
Ans.Detecting Overfitting:
    1.Training vs. Validation Curve:
        Plot training & validation accuracy over time. If the validation accuracy starts to decrease while
        training accuracy continues to increase,its  sign of overfitting.
    2.Cross-Validation:
        Split data into flods & train/test on each fold.If the model performs well on training data but poorly 
        on test data,its overfitting.
    3.Regularization Techniques:
        Apply regularization (e.g. L1,L2) & check if it improves model performance.
    4.Early Stopping:
        Stop training when validation performance starts to degrade.
    5.Model Complexity:
        Check if the model is too complex for the data.
        
    Detecting Underfitting:
    1.Training vs. Validation Curve:
        If the training accuracy is low & doesnt improve over time,its a sign of underfitting.
    2.Model Complexity:
        Check if the model is too simple for the data.
    3.Feature Engineering:
        Check if relevant features are missing.
    4.Hyperparameter Tuning:
        Try increasing model capacity or adjusting Hyperparameters.
    5.Data Quality:
        Check for noisy or irrelevant data.
        
    Determining Overfitting or Underfitting:
        1.Check Model Performance Metrics:
            Evaluate model performance on training & test data using metrics like accuracy,precision,recall,
            F1-score,etc.
        2.Visualize Model Predictions:
            Plot predicted values against actual values to check for patterns.
        3.Use Learning Curves:
            Plot training & validation accuracy over time to identify overfitting or underfitting.
        4.Use Cross-Validation:
            Split data into flods & train/test on each flod to evaluate model performance.
        5.Monitor Model Complexity:
            Check model complexity & adjust accordingly.
```
