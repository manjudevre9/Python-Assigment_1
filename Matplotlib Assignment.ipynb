```python
Que.2.What is scatter plot? Use the following code to generate data for x & y.Using this generated data plot a 
      scatter plot.
Ans.A scatter plot is a type of plot that displays the relationship between two variables by plotting points 
      on a coordinate plane.    
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x=3+np.random.normal(0,2,50)
y=3+np.random.normal(0,2,len(x))

plt.scatter(x,y)

plt.tital('Scatter Plot Example')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
Que.3.Why is the subplot() function used? Drow four line plots using the subplot() function.
Ans.It is used to create multiple plots in a single figure.
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
y3=np.sin(2*x)
y4=np.cos(2*x)

fig,axs=plt.subplots(2,2)

axs[0,0].plot(x,y1)
axs[0,0].set_title('Sine_Wave')

axs[0,1].plot(x,y2)
axs[0,1].set_title('Cosine_Wave')

axs[1,0].plot(x,y3)
axs[1,0].set_title('Double Frequency Sine')

axs[1,1].plot(x,y4)
axs[1,1].set_title('Double Frequency Cosine')

fig.tight_layout()

plt.show()

```

```python
Que.4.What is bar plot?Whay is it used? Using the following data plot a bar plot & a horizontal bar plot.
Ans.A bar plot,also known as a bar chart or bar graph,is a type of graphical display used to compare the values
    of different categories or groups.
It consist of:
    1.Bars.
    2.X-axis.
    3.Y-axis.
```

```python
# Bar plot:`
import numpy as np
 
company=np.array(["Apple","Microsoft","Google","AMD"])
profit=np.array([3000,8000,1000,10000])

plt.bar(company,profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Bar Plot")

plt.show()
```

```python
# Horizontal Bar Plot:
import numpy as np

company=np.array(["Apple","Microsoft","Google","AMD"])
profit=np.array([3000,8000,1000,10000])

plt.bar(company,profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Horizontal Bar Plot")

plt.show()
```

```python
Que.5.What is box plot?Why is it used?Using the following data plot a box plot.
Ans.Box Plot: Abox plot ,also known as a box & whisker plot,is a graphical representation of a dataset that 
              displays the distribution of values in a clear & concise way,it is used to:
        1.Show the central tendency.
        2.Display the variability.
        3.Identify outliers.
A box plot consists of:
    1.Box.
    2.Median.
    3.Whiskers.
    4.Outliers.
Box plots are useful for:
    1.Comparing distributions.
    2.Identifying skewness.
    3.Detecting outliers.
    4.Summarizing large datasets.
Common applications of box plots includes:
    1.Quality control.
    2.Finance.
    3.Biology.
    4.Social sciences.
```

```python
import numpy as np

box1=np.random.normal(100,10,200)
box2=np.random.normal(90,20,200)

plt.boxplot(box1)
plt.boxplot(box2)
plt.title("Box Plot")

plt.show()
```
